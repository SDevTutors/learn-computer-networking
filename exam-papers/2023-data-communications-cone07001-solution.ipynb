### Question 1.A


A sinusoidal signal $x(t)$ is described by the equation:

$$x(t) = -4.25 \sin(2750t + 1.6)$$

1. What is the frequency of the signal?
2. Sketch the signal for $0 \leq t \leq 3$ msec.  
3. If the signal was delayed by 0.5 msec, write out the new equation for the signal.

**[8 Marks]**


#### Answer 1.A.1

The frequency of the signal is the coefficient of $t$ in the argument of the sine function. Therefore, the frequency of the signal is $f = \frac{2750}{2\pi} \approx 437.5$ Hz.

#### Answer 1.A.2

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
amplitude = -4.25
frequency = 2750 / (2 * np.pi)  # Convert angular frequency to Hz
phase_shift = 1.6
time = np.linspace(0, 3e-3, 1000)  # Time vector from 0 to 3 milliseconds

# Define the signal
x_t = amplitude * np.sin(2750 * time + phase_shift)

# Plot the signal
plt.figure()
plt.plot(time * 1e3, x_t)  # Convert time to milliseconds for plotting
plt.title('Sinusoidal Signal $x(t) = -4.25 \sin(2750t + 1.6)$')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.ylim(-5, 5)
plt.xlim(0, 3)
plt.show()
```

#### Answer 1.A.3
If the signal was delayed by 0.5 msec, write out the new equation for the signal.

To Add a delay to the signal, we need to subtract the delay from the phase term of the signal ($1.6$). Therefore, the new equation for the signal is:

$x(t) = -4.25 \sin(2750(t) + 1.6 - 0.5) =$  
$x(t) = -4.25 \sin(2750(t) + 1.1)$

$$\boxed{ x(t) = -4.25 \sin(2750(t) + 1.1) }$$


 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
amplitude = -4.25
frequency = 2750 / (2 * np.pi)  # Convert angular frequency to Hz
phase_shift = 1.6
time = np.linspace(0, 3e-3, 1000)  # Time vector from 0 to 3 milliseconds

# Define the original signal
x_t = amplitude * np.sin(2750 * time + phase_shift)

# Define the new signal with the time shift
x_t_shifted = amplitude * np.sin(2750 * (time - 0.5e-3) + phase_shift)

# Plot the signals
plt.figure()
plt.plot(time * 1e3, x_t, label='$x(t) = -4.25 \sin(2750t + 1.6)$')
plt.plot(time * 1e3, x_t_shifted, label='$x(t) = -4.25 \sin(2750(t - 0.5) + 1.6)$', linestyle='--')
plt.title('Sinusoidal Signals')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.ylim(-5, 5)
plt.xlim(0, 3)
plt.legend()
plt.show()
```

### Question 1.B

Signals can be viewed in the time domain or the frequency domain.

#### Question 1.B.1

Explain the concepts of time domain and frequency domain.  

#### Answer 1.B.1

### Time Domain vs. Frequency Domain

**Time Domain:**  
The time domain represents a signal as a function of time. When we analyze a signal in the time domain, we observe how the signal's amplitude or value changes over time. This is the most intuitive representation of a signal, often visualized as a waveform where the x-axis represents time, and the y-axis represents the amplitude (or value) of the signal.

- **Key Characteristics:**
  - Focuses on the behavior of the signal over time.
  - Amplitude changes are tracked as time progresses.
  - Commonly used for signals like audio, video, and electrical waveforms.

- **Example:** 
  An audio signal might be represented as a continuous waveform showing how the sound intensity changes from one moment to the next over time.

**Frequency Domain:**  
The frequency domain represents a signal in terms of its frequency components rather than time. When analyzing a signal in the frequency domain, we look at how much of the signal's energy is present at each frequency. This representation is achieved through mathematical transformations like the Fourier Transform, which decomposes the signal into its constituent sinusoidal frequencies.

- **Key Characteristics:**
  - Focuses on the distribution of the signal's energy across different frequencies.
  - Shows which frequencies are present in the signal and how strong they are.
  - Useful for understanding periodicity, filtering, and analyzing bandwidth.

- **Example:** 
  The same audio signal, when transformed into the frequency domain, would show the dominant frequencies (such as bass, midrange, and treble) and their intensities, instead of how the signal changes over time.

**Comparison:**  
- **Time Domain:** Emphasizes the progression and changes of the signal as time moves forward.
- **Frequency Domain:** Highlights the underlying frequency components of the signal, providing insight into periodicities and spectral content.

Both domains provide complementary views of the same signal, and one can often switch between them using mathematical tools like the Fourier Transform (for time to frequency) or the Inverse Fourier Transform (for frequency to time).

### Question 1.D (5 Marks)


Sampling an analogue signal to convert it to a digital signal sometimes results in aliasing. Explain how aliasing occurs and how it can be prevented.


### Answer 1.D


Aliasing occurs when the sampling rate is too low to capture the frequency content of the signal. This results in the signal being reconstructed with a lower frequency than the original signal. Aliasing can be prevented by using a sampling rate that is at least twice the highest frequency component of the signal. This is known as the Nyquist-Shannon sampling theorem.

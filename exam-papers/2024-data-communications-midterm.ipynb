### Question 1.B

Signals can be viewed in the time domain or the frequency domain.

#### Question 1.B.1

Explain the concepts of time domain and frequency domain.  

#### Answer 1.B.1

**Time Domain vs. Frequency Domain:**  

**Time Domain:**  
The time domain represents a signal as a function of time. When we analyze a signal in the time domain, we observe how the signal's amplitude or value changes over time. This is the most intuitive representation of a signal, often visualized as a waveform where the x-axis represents time, and the y-axis represents the amplitude (or value) of the signal.

**Key Characteristics:**
- Focuses on the behavior of the signal over time.
- Amplitude changes are tracked as time progresses.
- Commonly used for signals like audio, video, and electrical waveforms.

**Example:** 
An audio signal might be represented as a continuous waveform showing how the sound intensity changes from one moment to the next over time.

**Frequency Domain:**  
The frequency domain represents a signal in terms of its frequency components rather than time. When analyzing a signal in the frequency domain, we look at how much of the signal's energy is present at each frequency. This representation is achieved through mathematical transformations like the **Fourier Transform**, which decomposes the signal into its constituent sinusoidal frequencies.

**Key Characteristics:**  
- Focuses on the distribution of the signal's energy across different frequencies.
- Shows which frequencies are present in the signal and how strong they are.
- Useful for understanding periodicity, filtering, and analyzing bandwidth.

- **Example:** 
  The same audio signal, when transformed into the frequency domain, would show the dominant frequencies (such as bass, midrange, and treble) and their intensities, instead of how the signal changes over time.

**Comparison:**  
- **Time Domain:** Emphasizes the progression and changes of the signal as time moves forward.
- **Frequency Domain:** Highlights the underlying frequency components of the signal, providing insight into periodicities and spectral content.

Both domains provide complementary views of the same signal, and one can often switch between them using mathematical tools like the Fourier Transform (for time to frequency) or the Inverse Fourier Transform (for frequency to time).

#### Question 1.B.2 (6 Marks)

Explain how a frequency domain view could be generated from a time series representation of a signal.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Define the parameters
amplitude = 1.0
frequency = 1.0  # Frequency in Hz
sampling_rate = 100.0  # Sampling rate in Hz
duration = 2.0  # Duration in seconds

# Generate the time vector
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Generate the signal in the time domain
x_t = amplitude * np.sin(2 * np.pi * frequency * t)

# Compute the Fourier Transform to get the frequency domain representation
x_f = fft(x_t)
frequencies = fftfreq(len(t), 1 / sampling_rate)

# Plot the signal in the time domain
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, x_t)
plt.title('Time Domain')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Plot the signal in the frequency domain
plt.subplot(2, 1, 2)
plt.plot(frequencies[:len(frequencies)//2], np.abs(x_f)[:len(frequencies)//2])
plt.title('Frequency Domain')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

### Question 2.D (6 Marks)


A data transmission system transmitter launches +22 dBm of power into a cable. The cable is 15 km in length and has a loss of 0.8 dB per kilometre. What is the received power level in watts at the receiver?


### Answer 2.D

To calculate the received power at the receiver in watts, we can follow these steps:

**Step 1: Determine the total loss in dB**:

- The cable loss is 0.8 dB per kilometer.
- The cable is 15 km long.

The total loss $L_{\text{total}}$ in dB is given by:

$$L_{\text{total}} = \alpha \times d$$

Where:
- $L_{\text{total}}$ is the total loss in dB,
- $\alpha$ is the attenuation (i.e *loss*) per unit distance (in dB per kilometer, for example) = 0.8 dB/km,
- $d$ is the total length of the transmission medium (in kilometers, meters, etc., depending on the units of $\alpha$) = 15.

   $L_{\text{total}} = \alpha \times d$  
   
   $L_{\text{total}} = 0.8 \, \text{dB/km} \times 15 \, \text{km} = 12 \, \text{dB}$

**Step 2: Calculate the received power in dBm**:

- The transmitter launches +22 dBm of power.
- The cable introduces a loss of 12 dB.

The received power $P_{\text{received}}$ in dBm is:

$P_{\text{received}} = P_{\text{transmitted}} - L_{\text{total}}$  

$P_{\text{received}} = 22 \, \text{dBm} - 12 \, \text{dB} = 10 \, \text{dBm}$  

**Step 3: Convert the received power from dBm to watts**:

To convert power from dBm to watts, use the following formula:

Note: $\times 10^{-3}$ is used to convert milliwatts to watts.

$$P_{\text{watts}} = 10^{(\frac{P_{\text{dBm}}}{10})} \times 10^{-3}$$  

Substituting the value of $P_{\text{dBm}} = 10$:

$P_{\text{watts}} = 10^{(\frac{10}{10})} \times 10^{-3} = 10^1 \times 10^{-3} = 10 \times 10^{-3} = 0.01 \, \text{W}$  

### **Final Answer:**
The received power at the receiver is **0.01 watts** or **10 milliwatts**.

Outline 3 advantages of using modulation in a data communication system.



### Answer 3.A



Here are three key advantages of using modulation in a data communication system:

1. **Efficient Use of Bandwidth:**
   - **Advantage:** Modulation allows multiple signals to be transmitted over a shared medium by shifting each signal to a different frequency range. This is commonly referred to as **frequency division multiplexing (FDM)**.
   - **Benefit:** It enables the simultaneous transmission of multiple data streams, maximizing the available bandwidth and improving the efficiency of the communication system.
2. **Improved Signal Transmission Over Long Distances:**
   - **Advantage:** Modulation, particularly at higher frequencies, helps in transmitting signals over longer distances with less degradation. Low-frequency signals (such as baseband signals) suffer more from attenuation and interference, while modulated signals in higher frequency bands (such as radio waves) are more resilient.
   - **Benefit:** It extends the transmission range and improves signal clarity, making communication over long distances more reliable.
3. **Reduced Interference and Noise:**
   - **Advantage:** By shifting the frequency of the baseband signal to a higher frequency range, modulation can avoid the effects of noise and interference that are typically concentrated at lower frequencies. Additionally, modulation schemes such as frequency modulation (FM) and phase modulation (PM) are inherently more resistant to certain types of noise.
   - **Benefit:** This results in better signal quality, with less distortion and fewer errors during transmission.

In summary, modulation improves bandwidth efficiency, enhances long-distance communication, and reduces the impact of noise and interference in data communication systems.

### Question 3.B (6 Marks)


Digital modulation systems are often less susceptible to noise than analogue modulation systems. Explain.


### Answer 3.B


Digital modulation systems are less susceptible to noise compared to analog modulation systems for several reasons:

1. **Discrete Signal Levels and Threshold Detection**  
Digital signals consist of discrete levels, often just two (representing binary 0s and 1s), which makes it easier for a receiver to distinguish between the signal and noise. Small amounts of noise may slightly distort the signal, but as long as the distortion doesn't push the signal past a certain threshold, the receiver can still correctly interpret the data.

2. **Error Detection and Correction Techniques**  
Digital modulation schemes often incorporate error detection and correction techniques, such as parity checks or more sophisticated algorithms like forward error correction. These techniques help identify and correct errors introduced by noise, further reducing its impact on communication reliability.

3. **Noise-Resilient Coding and Interleaving**  
Digital modulation can utilize techniques such as coding and interleaving, which spread the data across multiple symbols or time slots. This makes the system more robust against noise and interference, as the effect of noise on any one symbol is mitigated by the ability to reconstruct the signal using the information spread across the system.

In summary, the discrete nature of digital signals, combined with error correction techniques and noise-resilient encoding methods, makes digital modulation systems inherently more resistant to noise than their analog counterparts.

### Question 3.C (7 Marks)

A data communications receiver has an SNR of 35 dB and a bandwidth of 1.5 MHz. How many levels should it use to get the maximum bitrate? Suggest a suitable modulation scheme to achieve the maximum bitrate.

### Answer 3.C

To determine the number of levels required to achieve the maximum bitrate, we can use the Shannon-Hartley theorem, which relates the maximum achievable data rate to the signal-to-noise ratio (SNR) and the bandwidth of the channel.

The Shannon-Hartley theorem is given by:

$$C = B \log_2(1 + \text{SNR}_{\text{linear}})$$

Where:

- $C$ is the channel capacity (maximum achievable data rate),
- $B$ is the bandwidth of the channel,
- $\text{SNR}_{\text{linear}}$ is the signal-to-noise ratio in linear scale in Hertz (H).

Given:
- $B = 1.5MHz = 1.5 \times 10^6$ Hz.
- $\text{SNR} = 35$ dB,

First, we need to convert the SNR from dB to a linear scale:

$\text{SNR}_{\text{linear}} = 10^{\left(\frac{\text{SNR}}{10}\right)} = 10^{\left(\frac{35}{10}\right)} = 10^{3.5} \approx 3,162.3$  

Substitute the values into the Shannon-Hartley theorem:

$B \log_2(1 + \text{SNR}_{\text{linear}}) = C$  
$(1.5 \times 10^6) \log_2(1 + 3162.3) =$  
$(1.5 \times 10^6) \log_2(3163.3) \approx$  
$(1.5 \times 10^6) \times 11.62 \approx$  
$17.43 \times 10^6 \text{ bps} = $  
$17.43 \text{ Mbps}$


**Modulation Scheme and Levels:**  
To determine the number of levels in a digital modulation scheme (such as Quadrature Amplitude Modulation, QAM), the formula relating the number of levels $M$ to the bitrate $R$, bandwidth $B$, and bits per symbol $\log_2(M)$ is:

$R = B \cdot \log_2(M)$

Since we have already calculated that the maximum bitrate is 17.43 Mbps, and the bandwidth is 1.5 MHz, we can determine the required number of levels $M$ by solving for $\log_2(M)$:

*Keep M as an expression of $2^n$ where n is the number of bits per symbol*

Let

- $R = 17.43 \times 10^6$ bps,
- $B = 1.5MHz = 1.5 \times 10^6$ Hz.

$R = B \cdot \log_2(M)$  

$17.43 \times 10^6 = 1.5 \times 10^6 \times \log_2(M)$  
$17.43 = 1.5 \times \log_2(M)$  
$\dfrac{17.43}{1.5} = \log_2(M)$  
$\log_2(M) = 11.62$

Applying Logarithmic properties:

$Log_a(b) = c \implies a^c = b$  

Thus, the required number of levels is:

$\log_2(M) = 11.62 \implies M = 2^{11.62}$

You cannot have a fractional number of levels, so you would round up to the nearest integer.

$\therefore M \approx 2^{12} = 4,096$

The closest suitable modulation scheme would be **4096-QAM** (which uses 12 bits per symbol, as $\log_2(4096) = 12$).

**Final Answer:**  

To achieve the maximum bitrate with an SNR of 35 dB and a bandwidth of 1.5 MHz, the receiver should use a modulation scheme that supports approximately **17.43 Mbps**. A suitable modulation scheme would be **4096-QAM**, which uses **12 bits per symbol** and supports a high data rate.

QPSK with a peak amplitude value of 1V and phases of 0 degrees, 90 degrees, 180 degrees, and 270 degrees.


```python
import numpy as np
import matplotlib.pyplot as plt

# QPSK parameters
amplitude_qpsk = 1
phases_qpsk = [0, np.pi/2, np.pi, 3*np.pi/2]
labels_qpsk = ['0Â°', '90Â°', '180Â°', '270Â°']

# Calculate QPSK constellation points
qpsk_points = [amplitude_qpsk * np.exp(1j * phase) for phase in phases_qpsk]

# Plot QPSK constellation diagram
plt.figure(figsize=(6, 6))
plt.scatter(np.real(qpsk_points), np.imag(qpsk_points), color='blue')
for point, label in zip(qpsk_points, labels_qpsk):
    plt.text(np.real(point), np.imag(point), f'{label}\n({np.real(point):.1f}, {np.imag(point):.1f})', fontsize=12, ha='center')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('QPSK Constellation Diagram')
plt.xlabel('In-phase (I)')
plt.ylabel('Quadrature (Q)')
plt.grid(True)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

       Q = 4 \sin(315^\circ) = 4 \cdot -\frac{\sqrt{2}}{2} = -2\sqrt{2} \approx -2.83
       $$
       Point: $(2.83, -2.83)$

### Summary of Points:
- **Amplitude = 2V**:
  - $(1.41, 1.41)$
  - $(-1.41, 1.41)$
  - $(-1.41, -1.41)$
  - $(1.41, -1.41)$

- **Amplitude = 4V**:
  - $(2.83, 2.83)$
  - $(-2.83, 2.83)$
  - $(-2.83, -2.83)$
  - $(2.83, -2.83)$

These are the points you would plot on the I-Q plane to draw the 8-QAM constellation diagram by hand.

```python
import numpy as np
import matplotlib.pyplot as plt

# 8-QAM parameters
amplitudes_8qam = [2, 4]
phases_8qam = [np.pi/4, 3*np.pi/4, 5*np.pi/4, 7*np.pi/4]
labels_8qam = ['45Â°', '135Â°', '225Â°', '315Â°']

# Calculate 8-QAM constellation points
qam_points = [amplitude * np.exp(1j * phase) for amplitude in amplitudes_8qam for phase in phases_8qam]

# Plot 8-QAM constellation diagram
plt.figure(figsize=(6, 6))
plt.scatter(np.real(qam_points), np.imag(qam_points), color='red')
for i, point in enumerate(qam_points):
    label = labels_8qam[i % len(labels_8qam)]
    plt.text(np.real(point), np.imag(point), f'{label}\n({np.real(point):.1f}, {np.imag(point):.1f})', fontsize=12, ha='center')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('8-QAM Constellation Diagram')
plt.xlabel('In-phase (I)')
plt.ylabel('Quadrature (Q)')
plt.grid(True)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
